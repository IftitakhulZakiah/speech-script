# D. Preparing for lexicon (unique words)

```python
version = 'v.0.2.5'
```

```python
# file = open(version + "/text.txt", "r")
# file = open('text/3-sister.txt', 'r')
file = open('teks-rm-adapt.txt', 'r')

all_text = ''
for line in file:
    idx = line.split(' ')[0] # get the id audio
    text = line.replace(idx, '').strip().lower() + ' '
    all_text += text
    
all_text
```

```python
tokens = all_text.split(' ')
unique_tokens = set(tokens)
len(unique_tokens)
```

```python
unique_tokens
```

```python
# split token that have char besided of a-z and dash
accepted_char = 'qwertyuiopasdfghjklzxcvbnm-'

accepted_tokens = []
rejected_tokens = []

for token in unique_tokens:
    isAccepted = True
    for char in token:
        if char not in accepted_char:
            isAccepted = False
    if isAccepted:
        accepted_tokens.append(token)
    else:
        rejected_tokens.append(token)
        
accepted_tokens.sort()
rejected_tokens.sort()
```

```python
print(len(rejected_tokens), len(accepted_tokens))
```

```python
# files = open(version + '/listword.txt', 'w')
# files = open('text/3-sister-listword.txt', 'w')
files = open('rm-listword.txt', 'w')
for token in accepted_tokens:
    files.write(token + '\n')
```

```python
# files = open(version + '/rejected_listword.txt', 'w')
# files = open('text/3-sister-rejected_listword.txt', 'w')
files = open('rm-rejected_listword.txt', 'w')
for token in rejected_tokens:
    token = token.replace('<','')
    files.write(token.replace('>', '') + '\n')
```

```python
################################################################# Insert the <>

file = open('rm-rejected_lexicon.txt', 'r')

tokens_ori = []
for line in file:
    token = line.split(' ')[0]
    tokens_ori.append(line.replace(token, '<' + token + '>'))
    
file.close()
```

```python
# file = open(version + '/rejected_listword_lex2.txt', 'w')
file = open('rm-rejected_lexicon2.txt', 'w')

for token in tokens_ori:
    file.write(token)
```
